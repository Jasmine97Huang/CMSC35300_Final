```python
def main():
    total_match_pair = []
    for word in tqdm.tqdm(same_vocab):
        match_pair = Pair_Word(word,ppmi,svd_ppmi,vocab_list_lyrics,sim_word_pair)
        if match_pair != ([], []):
            total_match_pair+=match_pair[0]
            total_match_pair+=match_pair[1]
            #print(match_pair)
    
    with open('Total_match_words.pickle','wb') as handle:
        pickle.dump(total_match_pair, handle, protocol=pickle.HIGHEST_PROTOCOL)
    return total_match_pair
```

```python
total_match_pair = main()
```
